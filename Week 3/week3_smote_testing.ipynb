## Load Dataset

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Shilpa S Nair\Desktop\PredictiveMaintenance_Shilpa\Project1_Predictive-Maintenance\data\engineered_iot_weather_dataset (1).csv")
```

## Prepare features and target

```python
leakage_cols = [
    "TWF",
    "HDF",
    "PWF",
    "OSF",
    "RNF"
]

id_cols = [
    "UDI",
    "Product ID",
    "date"
]

target = "Machine failure"

X = df.drop(
    columns=leakage_cols + id_cols + [target]
)

X = pd.get_dummies(
    X,
    columns=["Type"],
    drop_first=True
)

y = df[target]
```

## Split the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.2,

    stratify=y,

    random_state=42
)
```

## SMOTE function

```python
import smote_balancer
```

```python
from smote_balancer import apply_smote
```

## apply smote

```python
X_train_smote, y_train_smote = apply_smote(
    X_train,
    y_train
)
```

## Compare class distributions

Before SMOTE:

```python
print("Before SMOTE")
print(y_train.value_counts())
```

After SMOTE:

```python
print("\nAfter SMOTE")
print(y_train_smote.value_counts())
```

## Compare shapes

```python
print("Original Training Shape:", X_train.shape)
print("Balanced Training Shape:", X_train_smote.shape)
```
